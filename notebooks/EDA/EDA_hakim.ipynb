```python
calendar = pd.read_csv('data/timetables/gtfs3Sept/calendar.txt', sep = ",")
calendar_dates = pd.read_csv('data/timetables/gtfs3Sept/calendar_dates.txt', sep = ",")
```

```python
calendar['start_date'] = calendar['start_date'].apply(lambda x: pd.to_datetime(str(x), format = '%Y%m%d'))
calendar['end_date'] = calendar['end_date'].apply(lambda x: pd.to_datetime(str(x), format = '%Y%m%d'))
calendar_dates['date'] = calendar_dates['date'].apply(lambda x: pd.to_datetime(str(x), format = '%Y%m%d'))
```
